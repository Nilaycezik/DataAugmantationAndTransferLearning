##### Data Augmantation Algoritma modellerinden olan Keras Sequential modelinin uygulanması ve kullanılması

##### Giriş katmanı

```python
model = Sequential()
model.add(Dense(30, input_dim=14, activation='tanh'))
```

```python
model.add(Dense(20, activation='tanh'))
```

##### Çıkış katmanı

```python
model.add(Dense(1, activation='sigmoid'))
```

##### Adam algoritması birinci dereceden gradyan bazlı optimizasyon için iyi bir algoritma olduğu için kullanılmıştır.

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(Input_train, Target_train, epochs=10, verbose=1)
```

##### Başarım Hesaplanması

```python
model.summary()
score = model.evaluate(Input_test, Target_test, verbose=0)
print('Model Accuracy = ',score[1])
```

#### VERİ SETİ VE ÖZELLİKLERİ
Bu veritabanı toplamda 76 öznitelik içerir, ancak bunların 14'ünün alt kümesini kullanmaktadır. Özellikle, Cleveland veri tabanı, bu tarihe kadar ML araştırmacıları tarafından kullanılan tek veri tabanıdır. "Target" alanı, hastada kalp hastalığının varlığını ifade eder. 0 (hastalık yok) ile 4 arasında değer alan bir tamsayıdır.  
1.	age
2.	sex
3.	chest pain type (4 values)
4.	resting blood pressure
5.	Serum cholestoral in mg/dl
6.	fasting blood sugar > 120 mg/dl
7.	resting electrocardiographic results (values 0,1,2)
8.	maximum heart rate achieved
9.	exercise induced angina
10.	oldpeak = ST depression induced by exercise relative to rest
11.	the slope of the peak exercise ST segment
12.	number of major vessels (0-3) colored by flourosopy
13.	thal: 3 = normal; 6 = fixed defect; 7 = reversable defect

#### BAŞARIM SONUÇLARI 
Data augmantation algoritma modellerinden olan Keras Sequential modeli kullanılarak elde edilen sonuçlar aşağıda verilmiştir.
İlk eğitim için parametre değerleri giriş katmanı için nöron sayısı 30 aktivasyon parametresi tanh seçilmiştir. Bu eğitimde bir tane ara katman kullanılmış olup 20 nörondan oluşmaktadır ve aktivasyon parametresi bu katman için de tanh seçilmiştir. 1 tane çıkışı bulunan çıkış katmanında ise aktivasyon parametresi sigmoid seçilmiştir. Elde edilen bu girdi ve çıktı değerlerine karşılık optimizasyon için adam algoritması seçilmiştir. Bunun sebebi birinci dereceden gradyan bazlı optimizasyon için iyi bir algoritma olduğu içindir. 10 iterasyon sayısı kullanılmıştır. Yukarıda verilen parametrelerden oluşan eğitim seti için elde edilen başarım oranı 0.7033 olarak elde edilmiştir.
İkinci eğitim için parametre değerleri giriş katmanı için nöron sayısı 20 aktivasyon parametresi sigmoid seçilmiştir. Bu eğitimde bir tane ara katman kullanılmış olup 10 nörondan oluşmaktadır ve aktivasyon parametresi bu katman için de sigmoid seçilmiştir. 5 tane çıkışı bulunan çıkış katmanında ise aktivasyon parametresi tanh seçilmiştir. Elde edilen bu girdi ve çıktı değerlerine karşılık optimizasyon için adam algoritması seçilmiştir. 10 iterasyon sayısı kullanılmıştır.Yani eğitim seti 10. epoch da sonlanmıştır. Yukarıda verilen parametrelerden oluşan eğitim seti için elde edilen başarım oranı 0.5385 olarak elde edilmiştir.
Üçüncü eğitim için parametre değerleri giriş katmanı için nöron sayısı 40 aktivasyon parametresi tanh seçilmiştir. Bu eğitimde iki tane ara katman kullanılmış olup bir tanesi 20, diğeri 10 nörondan oluşmaktadır ve aktivasyon parametresi ilk gizli (ara) katman için tanh seçilirken, ikinci gizli (ara) katman için sigmoid seçilmiştir. 10tane çıkışı bulunan çıkış katmanında ise aktivasyon parametresi tanh seçilmiştir.
